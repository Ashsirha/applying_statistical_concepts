### Clustering and Resampling

Let's set up our workspace and use the **Iris dataset** from `scikit-learn`. This dataset is a classic dataset in machine learning and statistics, widely used for clustering tasks. It consists of many samples of iris flowers. Here are the key features and characteristics of the dataset:

##### Features:
1. **Sepal Length**: The length of the sepal in centimeters.
2. **Sepal Width**: The width of the sepal in centimeters.
3. **Petal Length**: The length of the petal in centimeters.
4. **Petal Width**: The width of the petal in centimeters.

```python
# Import standard libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

#### **Question 1:** 
#### Data inspection

#### Load the Iris dataset:

Use scikit-learn to load the Iris dataset and convert it into a Pandas DataFrame.
Display the first few rows of the dataset. How many observations (rows) and features (columns) does the dataset contain?

```python
from sklearn.datasets import load_iris
# Load the Iris dataset
iris_data = load_iris()

# Convert to DataFrame
iris_df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)

# Display the DataFrame
print(iris_df)

# Display the number of observations (rows) and features (columns)
num_rows, num_columns = iris_df.shape
print(f"The Iris dataset contains {num_rows} observations (rows) and {num_columns} features (columns).")
```

#### **Question 2:** 
#### Data-visualization

Let's create plots to visualize the relationships between the features (sepal length, sepal width, petal length, petal width).


```python
def plot_feature_pairs(data, feature_names, color_labels=None, title_prefix=''):
    """
    Helper function to create scatter plots for all possible pairs of features.
    
    Parameters:
    - data: DataFrame containing the features to be plotted.
    - feature_names: List of feature names to be used in plotting.
    - color_labels: Optional. Cluster or class labels to color the scatter plots.
    - title_prefix: Optional. Prefix for plot titles to distinguish between different sets of plots.
    """
    # Create a figure for the scatter plots
    plt.figure(figsize=(12, 10))
    
    # Counter for subplot index
    plot_number = 1
    
    # Loop through each pair of features
    for i in range(len(feature_names)):
        for j in range(i + 1, len(feature_names)):
            plt.subplot(len(feature_names)-1, len(feature_names)-1, plot_number)
            
            # Scatter plot colored by labels if provided
            if color_labels is not None:
                plt.scatter(data[feature_names[i]], data[feature_names[j]], 
                            c=color_labels, cmap='viridis', alpha=0.7)
            else:
                plt.scatter(data[feature_names[i]], data[feature_names[j]], alpha=0.7)
            
            plt.xlabel(feature_names[i])
            plt.ylabel(feature_names[j])
            plt.title(f'{title_prefix}{feature_names[i]} vs {feature_names[j]}')
            
            # Increment the plot number
            plot_number += 1

    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Show the plot
    plt.show()

# Get feature names
feature_names = iris_df.columns

# Use the helper function to plot scatter plots without coloring by cluster labels
plot_feature_pairs(iris_df, feature_names, title_prefix='Original Data: ')
```

**Question:**
- Do you notice any patterns or relationships between the different features? How might these patterns help in distinguishing between different species?


From the scatter plots, we can observe the following patterns and relationships between the different features:

1. **Petal Length vs. Petal Width**: There is a clear separation in this plot. Some flowers have smaller petal lengths and widths, while others have larger petal lengths and widths. This separation can help in distinguishing different groups of flowers.

2. **Sepal Length vs. Sepal Width**: The separation is not as clear as in the petal dimensions, but there is still some distinction. Some flowers tend to have wider sepals compared to others.

3. **Petal Length vs. Sepal Length**: This plot also shows a good separation. Some flowers have shorter petal lengths and sepal lengths, while others have longer petal lengths and sepal lengths.

4. **Petal Width vs. Sepal Length**: Similar to the Petal Length vs. Sepal Length plot, there is a noticeable separation.

These patterns indicate that petal dimensions (length and width) are more effective in distinguishing between the groups compared to sepal dimensions. By analyzing these relationships, we can better understand the characteristics that differentiate the groups and use this information for classification tasks.


#### **Question 3:** 
#### Data cleaning

```python
# Initialize the StandardScaler
scaler = StandardScaler()

# Scale all the features in the dataset
scaled_features = scaler.fit_transform(iris_df)

# Create a new DataFrame with scaled features
scaled_iris_df = pd.DataFrame(scaled_features, columns=iris_data.feature_names)

# Display the first few rows of the scaled DataFrame
print(scaled_iris_df.head())
```

Why is it important to standardize the features of a dataset before applying clustering algorithms like K-Means? Discuss the implications of using unstandardized data in your analysis.

Standardizing the features of a dataset before applying clustering algorithms like K-Means is crucial for several reasons:

1. **Equal Weighting**: Standardization ensures that each feature contributes equally to the distance calculations. Without standardization, features with larger scales can dominate the distance metric, leading to biased clustering results.

2. **Improved Convergence**: K-Means relies on distance calculations to assign points to clusters. Standardized data helps the algorithm converge more quickly and accurately by providing a more uniform scale for all features.

3. **Interpretability**: Standardized features make it easier to interpret the clustering results, as the influence of each feature is balanced.

### Implications of Using Unstandardized Data:
- **Biased Clustering**: Features with larger scales can disproportionately influence the clustering process, leading to clusters that are more reflective of those features rather than the overall data structure.
- **Poor Performance**: The algorithm may perform poorly, with longer convergence times and potentially suboptimal clustering results.
- **Misleading Insights**: The resulting clusters may not accurately represent the underlying patterns in the data, leading to incorrect conclusions and insights.

In summary, standardizing the features ensures that the clustering algorithm treats all features equally, leading to more accurate and meaningful clustering results.

#### **Question 4:** 
#### K-means clustering 

Apply the K-Means clustering algorithm to the Iris dataset. Choose the value 3 for the number of clusters (`k=3`) and fit the model. Assign cluster labels to the original data and add them as a new column in the DataFrame.

```python
# Apply the K-Means clustering algorithm to the Iris dataset
kmeans = KMeans(n_clusters=3, random_state=12345)
kmeans.fit(scaled_iris_df)

# Assign cluster labels to the original data and add them as a new column in the DataFrame
iris_df['Cluster'] = kmeans.labels_

# Display the DataFrame with cluster labels
print(iris_df)

# Use the helper function to plot scatter plots, colored by cluster labels
plot_feature_pairs(iris_df, feature_names, color_labels=iris_df['Cluster'], title_prefix='Clustered Data: ')
```

We chose `k=3` for the number of clusters arbitrarily. However, in a real-world scenario, it is important to determine the optimal number of clusters using appropriate methods.

**Question**: What is one method commonly used to determine the optimal number of clusters in K-means clustering, and why is this method helpful?

One commonly used method to determine the optimal number of clusters in K-means clustering is the **Elbow Method**. This method involves plotting the Within-Cluster Sum of Squares (WSSD) against the number of clusters (k). The optimal number of clusters is identified at the "elbow point," where the rate of decrease in WSSD sharply slows down. This helps balance model complexity and fit.

#### **Question 5:** 
#### Bootstrapping 

Implement bootstrapping on the mean of Petal Width. Generate 10000 bootstrap samples, calculate the mean for each sample, and compute a 90% confidence interval.

```python
# Extract Petal Width column
petal_width = iris_df['petal width (cm)']

# Number of bootstrap samples
n_bootstrap_samples = 10000

# Array to store the mean of each bootstrap sample
bootstrap_means = np.zeros(n_bootstrap_samples)

# Generate bootstrap samples and calculate the mean for each sample
for i in range(n_bootstrap_samples):
    bootstrap_sample = np.random.choice(petal_width, size=len(petal_width), replace=True)
    bootstrap_means[i] = np.mean(bootstrap_sample)

# Calculate the mean of Petal Width
mean_petal_width = np.mean(bootstrap_means)

# Compute the 90% confidence interval
lower_bound = np.percentile(bootstrap_means, 5)
upper_bound = np.percentile(bootstrap_means, 95)

# Display the result
print(f"Mean of Petal Width: {mean_petal_width}")
print(f"90% Confidence Interval of Mean Petal Width: ({lower_bound}, {upper_bound})")
```

**Question:**
- Why do we use bootstrapping in this context? What does it help us understand about the mean?

Bootstrapping is used in this context to estimate the sampling distribution of the mean of Petal Width by resampling with replacement from the original data. This technique helps us understand the variability and stability of the mean estimate. By generating many bootstrap samples and calculating the mean for each, we can construct a confidence interval, providing insight into the range within which the true mean is likely to fall. This helps us assess the reliability and precision of our mean estimate.

**Question:**
- What is the purpose of calculating the confidence interval from the bootstrap samples? How does it help us interpret the variability of the estimate?

Calculating the confidence interval from bootstrap samples helps us estimate the range within which the true population parameter is likely to fall. It provides a measure of the precision and reliability of our estimate. By examining the confidence interval, we can assess the variability and stability of the estimate. This helps in understanding the uncertainty associated with the sample mean.

**Question:**

- Reflect on the variability observed in the bootstrapped means and discuss whether the mean of the Petal Width appears to be a stable and reliable estimate based on the confidence interval and the spread of the bootstrapped means.


Reflecting on the variability observed in the bootstrapped means, we can see that the mean of the Petal Width appears to be a stable and reliable estimate. The 90% confidence interval for the mean Petal Width is (1.0967, 1.3), which is relatively narrow. This indicates that the bootstrapped means do not vary widely and are clustered closely around the mean value of approximately 1.1991. The narrow confidence interval and the tight clustering of the bootstrapped means suggest that the mean Petal Width is a consistent and dependable estimate, with limited variability.
